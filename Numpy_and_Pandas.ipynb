# Numpy Array Operations:

```python
import numpy as np

# 1. Create a 1D NumPy array for Week 1
temperatures_w1 = np.array([22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9])

```

```python
# 2. Inspection and properties
print("Week 1 Temperatures:", temperatures_w1)
print("Shape:", temperatures_w1.shape)
print("Data type:", temperatures_w1.dtype)
print("Number of elements:", temperatures_w1.size)
```

```python
# 3. Array operations
# Convert Celsius to Fahrenheit
temperatures_f = (temperatures_w1 * 9/5) + 32
print("\nTemperatures in Fahrenheit:", temperatures_f)
```

```python
# Find maximum, minimum, and mean
print("Maximum temperature:", temperatures_w1.max())
print("Minimum temperature:", temperatures_w1.min())
print("Mean temperature:", temperatures_w1.mean())
```

```python
# 4. Array slicing and indexing
print("\nFirst three days:", temperatures_w1[:3])
print("Weekend temperatures:", temperatures_w1[-2:])
print("Middle three days:", temperatures_w1[2:5])
```

```python
# 5. Create a 2D NumPy array for two weeks
temperatures = np.array([
    [22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9],
    [19.2, 22.5, 21.3, 24.0, 23.5, 22.8, 20.1]
])

```

```python
# 6. Inspect and slice the 2D array
print("\n2D Temperature Array:")
print(temperatures)
print("Shape:", temperatures.shape)
print("Data type:", temperatures.dtype)
print("Total number of elements:", temperatures.size)

# Extract temperatures for each week
week1 = temperatures[0]
week2 = temperatures[1]
print("\nWeek 1 temperatures:", week1)
print("Week 2 temperatures:", week2)

# Extract weekend temperatures (last two days) for both weeks
weekends = temperatures[:, -2:]
print("\nWeekend temperatures for both weeks:")
print(weekends)
```

# Pandas Series:

```python
import pandas as pd

# 1. Creating Pandas Series

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=["Rank1", "Rank2", "Rank3", "Rank4", "Rank5"]
)

print("Marks Series:")
print(marks)
```

```python
# 2. Indexing and Slicing

# Access the mark of the 1st rank student using integer index
print("\nMark of 1st Rank Student:")
print(marks[0])

# Retrieve the marks of the top 3 ranks using loc
print("\nTop 3 Rank Marks (using loc):")
print(marks.loc["Rank1":"Rank3"])

# Retrieve the mark of the 3rd-ranked student using iloc
print("\nMark of 3rd Rank Student (using iloc):")
print(marks.iloc[2])

# Boolean mask: Marks greater than 90
print("\nMarks Greater Than 90:")
print(marks[marks > 90])

```

```python
# 3. Manipulating Series

# Modify the mark of the 1st rank student
marks["Rank1"] = 100

print("\nAfter Updating Rank1 Marks:")
print(marks)

# Remove the last rank entry
marks = marks.drop("Rank5")

print("\nAfter Removing Rank5:")
print(marks)

# Compute CGPA by dividing marks by 10
cgpa = marks / 10

print("\nCGPA:")
print(cgpa)
```

# Pandas DataFrame:

```python
import pandas as pd

# 1. Creating Pandas DataFrame

transactions = pd.DataFrame({
    "TransactionID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "ProductCategory": [
        "Electronics", "Clothing", "Electronics", "Furniture", "Clothing",
        "Electronics", "Furniture", "Clothing", "Furniture", "Electronics"
    ],
    "Region": [
        "North", "South", "North", "East", "West",
        "North", "East", "West", "South", "North"
    ],
    "Amount": [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]
})

print("Transactions DataFrame:")
df = pd.DataFrame(transactions)
print(transactions)
```

```python
# Group by 'Region' and find the mean amount for each region.

print(df.groupby("Region")["Amount"].mean())
```
